# Проект "Kiva"


Краудфандинговая платформа kiva.org ставит своей целью предоставление финансовых услуг наиболее нуждающимся и необеспеченным людям. Заёмщики оставляют заявку на сайте, доноры поддерживают её  финансово, на месте партнёр организации выдаёт кредит, который погашается из этих средств, которые донорам возвращаются уже заёмщиками через платформу.

Представленный компанией датасет за 2014-2017гг. содержит информацию о займах на различные проекты в 85 разных странах.

Проект призван найти взаимосвязи характеристик со скоростью набора средств, мы исследуем, связана ли доля заёмщиков-женщин с основной религией страны. Выделим локальных финансовых партнёров, которые доверяют платформе и выдают кредиты до того, как собраны средства на их возмещение. Проанализируем как зависит сектор заявки от индекса бедности страны и в каких странах Kiva недостаточно представлена.

#### Обработка пропусков
[Назад к содержанию](#section_id)

```python
#Для столбцов use, region и tags пропуски заполню значением "no data"
#Информация в эти столбцы вносится вероятно пользователями. Геоданных для вычилсения конкретного места так же нет.
df['use'] = df['use'].fillna('unknown')
df['region'] = df['region'].fillna('unknown')
df['tags'] = df['tags'].fillna('unknown')
df['borrower_genders'] = df['borrower_genders'].fillna('unknown')
```

```python
#Для coutnry_code проверим, где есть пропуски.
#В случае несоответствия кода страны ее названию приоритет отдается названию страны.
df.query('country_code.isna()')['country'].value_counts()
```

```python
#Используя данные Википедии пропуски заполню корректным кодом для Намибии
df['country_code'] = df['country_code'].fillna('NAM')
```

```python
#Пропуски в partner_id свидетельствуют о том, что возможно кредит был выдан напрямую, без посредников.
#Пропишу отличающееся от всех остальных значение, которое при надобности буду исключать из рассчетов, чтобы не искажало картину.
df['partner_id'] = df['partner_id'].fillna(9999)
```

```python
#Пустые значения индекса бедности решено заменить медианным значением.
df.query('mpi.isna()')['country'].value_counts()
```

```python
df['mpi'] = df['mpi'].fillna(df['mpi'].median())
```

#### Обработка столбца borrower_genders
[Назад к содержанию](#section_id)

- В столбе **borrower_genders** представлена информация о поле заемщиков.
- Если заемщиков несколько, то через запятую указан пол каждого заемщика.
- Текстовое определение неудобно.
- Решение - вывод следующих столбцов
  - **m_cnt** - кол-ов мужчин в заявке
  - **f_cnt** - количество женщин в заявке
  - **brw_cnt** - общее количество  замещиков
  - **f_perc** - процент женщин в заявке
  - **gen_cat** - категория заявки по полу:
    - *F* - женщин в заявке > 50%;
    - *FM* - женщин в заявке = 50%;
    - *M* - женщин в заявке < 50%;

```python
#Количество мужчин в заявке
df.insert(19,"m_cnt",(df['borrower_genders'].str.count('male')-df['borrower_genders'].str.count('female')))

#Количество женщин в заявке
df.insert(20,"f_cnt",(df['borrower_genders'].str.count('female')))

#Число заемщиков в заявке
df.insert(21,"brw_cnt",(df['m_cnt']+df['f_cnt']))

#Процент женщин в заявке
df.insert(22,"f_perc",(round((df['f_cnt']/df['brw_cnt'])*100)))

#Определение категории заявки по полу
def gen(percentage):
    if percentage < 50:
        return 'M'
    elif percentage == 50:
        return 'FM'
    elif percentage > 50:
        return 'F'
    return 'unknown'
df.insert(23,"gen_cat", df['f_perc'].apply(gen))
```

```python
#Рассчитываю скорость набора средств.
#Рассматривается разница между времением пубикации заявки и временем, когда сумма была собрана
df['fund_speed'] = df['funded_time'] - df['posted_time']

#Дополнительно рассчитываю время, когда деньги предоставлены раньше, чем была опубликована заявка
df['dib_speed'] = df['disbursed_time'] - df['posted_time']

#Отдельным столбцом преобразую полученную разницу в количество дней для удобства рассчетов.
df['fund_days'] = np.ceil((df['fund_speed'] / np.timedelta64(1, 'D')).astype(float))
df['dib_days'] = np.round((df['dib_speed'] / np.timedelta64(1, 'D')).astype(float))
#df.info()
```

#### Определение скорости сбора средств

Введу категориальное деление скорости сбора средств
- **A** - средства собраны за 1 день и быстрее;
- **B** - средства собраны в срок 1-7 дней;
- **C** - средства собраны в срок 7-30 дней;
- **D** - средства собраны в срок 30-180 дней;
- **E** - средства собраны в срок 180-365 дней;
- **F** - средства собраны в срок более 365 дней.

```python
def day_cat(day):
    if day <= 1:
        return 'A'
    elif 1 < day <= 7:
        return 'B'
    elif 7 < day <= 30:
        return 'C'
    elif 30 < day <= 180:
        return 'D'
    elif 180 < day <= 360:
        return 'E'
    elif 360 < day:
        return 'F'
    return 'error'
df.insert(24,"day_cat", (df['fund_days'].apply(day_cat)))
```

#### Категория индекса бедности
[Назад к содержанию](#section_id)

Введу категориальное деление стран по индексу бедности. Чем ближе к  единице, тем беднее:
- **A** - индекс равен 0,5 и выше;
- **B** - индекс от 0,4 включительно до 0,5;
- **C** - индекс от 0,3 включительно до 0,4;
- **D** - индекс от 0,2 включительно до 0,3;
- **E** - индекс от 0,1 включительно до 0,2;
- **F** - индекс меньше 0,1;

```python
# Разделю индекс бедности на категории
def mp_i(mpi):
    if mpi >= 0.5:
        return 'A'
    elif 0.4 <= mpi < 0.5:
        return 'B'
    elif 0.3 <= mpi < 0.4:
        return 'C'
    elif 0.2 <= mpi < 0.3:
        return 'D'
    elif 0.1 <= mpi < 0.2:
        return 'E'
    elif 0.1 > mpi:
        return 'F'
    return 'error'
df["mpi_cat"] = (df['mpi'].apply(mp_i))

# Так же обозначу негласную границу между старанами с индексом выше 0,1 и ниже, как более грубое разделение.
```

```python
# Посмотрим разделение стран по категориям бедности
ind = 'mpi_cat'
val = 'country'
agg = 'nunique'
sort_by = 'mpi_cat'
asc = True
tit = 'Распределение стран по категориям бедности'
x_tit='Каегория бедности'
y_tit='Количество стран'
x_ax = None

grph_1(df)
```
